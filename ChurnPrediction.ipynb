```python
import pandas as pd
import matplotlib.pyplot as plt

def cat_vs_churn_plotting(cat, ax):
    cat_type_cat = sorted(df2[cat].unique())  # Sort the categories
    cat_type_count = []

    for i in cat_type_cat:
        cat_type_not_exited_count = df2[(df2[cat] == i) & (df2['Exited'] == 0)].shape[0]
        cat_type_exited_count = df2[(df2[cat] == i) & (df2['Exited'] == 1)].shape[0]
        cat_type_count.append([i, cat_type_not_exited_count, cat_type_exited_count])

    cat_type_df = pd.DataFrame(cat_type_count, columns=[cat, 'Not Exited', 'Exited'])
    cat_type_df.set_index(cat).sort_index().plot(kind='bar', stacked=True, ax=ax)
    ax.set_title(f'Customer Exit Counts by {cat}')
    ax.set_xlabel(cat)
    ax.set_ylabel('Count')

    for i in range(len(cat_type_df)):
        ax.text(i - 0.15, cat_type_df['Not Exited'][i] + 0.1, str(cat_type_df['Not Exited'][i]), ha='center', va='bottom')
        ax.text(i + 0.15, cat_type_df['Not Exited'][i] + cat_type_df['Exited'][i] + 0.1, str(cat_type_df['Exited'][i]), ha='center', va='bottom')


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to handle NaN values in the dataset
def handle_nans(X): 
    nan_mask = np.isnan(X) 
    col_means = np.nanmean(X, axis=0)
    X[nan_mask] = np.take(col_means, np.where(nan_mask)[1]) 
    return X

# Function to calculate Euclidean distance
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))

# Function to initialize centroids by randomly selecting 'k' points from the dataset
def initialize_centroids(X, k):
    np.random.seed(42)  # for reproducibility
    random_indices = np.random.choice(X.shape[0], k, replace=False)
    centroids = X[random_indices]
    return centroids

# Function to assign each point to the nearest centroid
def assign_clusters(X, centroids):
    clusters = []
    for point in X:
        distances = [euclidean_distance(point, centroid) for centroid in centroids]
        assigned_cluster = np.argmin(distances)  # find index of nearest centroid
        clusters.append(assigned_cluster)
    return np.array(clusters)

# Function to update centroids based on the assigned clusters
def update_centroids(X, clusters, k):
    new_centroids = np.zeros((k, X.shape[1]))
    for i in range(k):
        points_in_cluster = X[clusters == i]
        if points_in_cluster.shape[0] > 0:
            new_centroids[i] = np.mean(points_in_cluster, axis=0)
    return new_centroids

# Function to implement K-Means clustering
def k_means(X, k, max_iterations=100):
    centroids = initialize_centroids(X, k)
    prev_centroids = np.zeros_like(centroids)
    clusters = np.zeros(X.shape[0])
    
    for i in range(max_iterations):
        # Assign points to the nearest centroid
        clusters = assign_clusters(X, centroids)
        
        # Update centroids
        centroids = update_centroids(X, clusters, k)
        
        # Check if centroids have converged (i.e., no change)
        if np.all(centroids == prev_centroids):
            break
        
        prev_centroids = centroids.copy()
    
    return centroids, clusters

X=X.to_numpy()
X=handle_nans(X)
centroids,clusters=k_means(X,2)
print("Centroid Coordinates\n",centroids)
print("Labels of Clusters\n",clusters)


```
